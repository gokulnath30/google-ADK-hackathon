```python
import pandas as pd

flood_inventory = pd.read_csv('../data/India_Flood_Inventory_v3.csv')
flood_inventory.head()
```

```python
def row_to_text(r):
    return f"""
Event: {r['Main Cause']}
Location: {r['Location']}
District: {r['Districts']}
State: {r['State']}
Date: {r['Start Date']} to {r['End Date']} ({r['Duration(Days)']} days)
Severity: {r['Severity']}
Human Fatality: {r['Human fatality']}
Human Injured: {r['Human injured']}
Area Affected: {r['Area Affected']}
Description: {r['Description of Casualties/injured']}
"""
data = [{"id": i, "text": row_to_text(r)} for i, r in flood_inventory.iterrows()]
data
```

```python
from vertexai.language_models import TextEmbeddingModel

model = TextEmbeddingModel.from_pretrained("text-embedding-005")

for item in data:
    item["embedding"] = model.get_embeddings([item["text"]])[0].values
data
```

```python
from google.cloud import aiplatform

vector_index = aiplatform.MatchingEngineIndex.create_tree_ah_index(
    display_name="disaster-vector-index",
    contents_delta_uri="gs://your_bucket/embeddings",
    dimensions=768
)
vector_index.wait()
vector_index.resource_name
```
